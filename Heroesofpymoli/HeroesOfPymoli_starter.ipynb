## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create bins for ages
bins = [-1, 9, 14, 19, 24, 29, 34, 39, purchase_data["Age"].max()]
# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice the data and place it into bins

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

purch_age = purchase_data.groupby(['Age Ranges'])
purch_count = purch_age['Purchase ID'].count()
avg_p_price = round(purch_age['Price'].mean(),2)
total_value = purch_age['Price'].sum()

avg_p_person = round(((total_value/purch_count)),2)
#Create Summary DF
summary_table4 = pd.DataFrame({"Purchase Count": purch_count, "Average Purchase Price": avg_p_price, 
                               "Total Purchase Value": total_value, "Avg Total Purchase per Person": avg_p_person})
#Format Table
summary_table4['Average Purchase Price']=summary_table4["Average Purchase Price"].map("${:.2f}".format)
summary_table4['Total Purchase Value']=summary_table4["Total Purchase Value"].map("${:.2f}".format)
summary_table4['Avg Total Purchase per Person']=summary_table4["Avg Total Purchase per Person"].map("${:.2f}".format)
summary_table4
```
